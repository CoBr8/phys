# Extract and write surface properties from 20$M_\odot$ stellar evolution model


## Initialize session

```python
%pylab 
from nugridpy import nugridse as mp
from nugridpy import mesa as ms

# default data location (try mirror if default
# is not available)
#data_dir="/tmp/nugrid_data"   

# mirror data locations (some mirrors 
# may not be available at some time)

# data_dir="/data/nugrid_cedar"  # may be default    
data_dir="/data/nugrid_rpod4"    
# data_dir="/data/nugrid_vos"    


ms.set_nugrid_path(data_dir)
mp.set_nugrid_path(data_dir)
```

## The MESA stellar evolution model
Initialise the 20 solar-mass Z=0.02 MESA stellar evolution model:

```python
s=ms.star_log(mass=20,Z=0.02)
```

What's happening at the surface of the star? ...

```python
ifig=102;close(ifig);figure(ifig)
s.hrd_new()
legend(loc='lower right').draw_frame(False)
```

```python
# these are the available columns
# s.cols
```

```python
things=['star_age','log_L','log_Teff','log_R','log_g']
writedata=[]
for thing in things:
    writedata.append(s.get(thing))
writedata = array(writedata)
```

```python
savetxt('M20star-surface.dat', writedata.T,header='20Msun surface properties as a function of time; \
\nCreated with notebook "20Msun-surfacedata.ipynb" on \nNuGrid server http://wendi1.phys.uvic.ca\n\
Cols:   star_age [yr], log_L [log10 L/Lsun],    log_Teff [log10 T/K],    log_R [log10 R/Rsun],    log_g [log g [cgs]]')
```
