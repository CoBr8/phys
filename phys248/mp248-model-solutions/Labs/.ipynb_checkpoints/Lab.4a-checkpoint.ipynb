#### Activity 2b:
Create a plot of the two-dimensional Gaussian function
$$
h(x,y) = \exp{-(x^2 + y^2)}
$$
for $x \in [-1,1.2]$ and $y \in [-1,1.5]$. 
Do both a 
* 3D surface plot, and a
* contour plot with labels.

```python
def h(x,y):
    return exp(-(x**2+y**2))
n = 50
x=linspace(-1.,1.2,n)
y=linspace(-1.,1.5,n)
#x
```

```python
xv,yv = meshgrid(x,y)
hxy = h(xv,yv)
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(xv,yv,hxy, rstride=1, cstride=1, cmap=cm.viridis)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```

#### Activity 3:
The [Gaussian function](https://en.wikipedia.org/wiki/Gaussian_function) is given by 
$$
f(x) = a e^{-\frac{(x-b)^2}{2c^2}}
$$

Create a function called `gauss` which takes as input `x,a=1,b=0,c=1` and returns the function value. Use this function to make a plot of the Gaussian function for $x \in [-3,5]$ for the default values `a=c=1` and `b=0`. Then add lines for the default but changing in turn
1. $a=0.5$
2. $b=1$
3. $c=0.5$

Provide a legend for each line, and of course axis labels as well. Use a different line style for each line. Briefly describe the meaning of each parameter.

```python
%pylab inline
import numpy as np
def gauss(x,a=1,b=0,c=1):
    gg = a*np.exp(-(x-b)**2 / (2*c**2))
    return gg
x = linspace(-3,5,50)
plot(x,gauss(x),'--',label='a=b=c=1')
plot(x,gauss(x,a=0.5),'-',label='a=0.5')
plot(x,gauss(x,b=1),':',label='b=1')
plot(x,gauss(x,c=0.5),'-.',label='c=0.5')
xlabel('x')
ylabel('g(x)')
legend()
```

```python
import astropy.units as u
Area = 260000 * u.km**2
PowerperunitArea = 1050 * u.W/u.m**2
PowerperCustomer = 10399 * u.kW * u.hour
eta = 0.2
num_cust = 150e6
```

```python
Esol = Area * PowerperunitArea * eta * u.year
print(Esol.to('J'))
```

```python
print(num_cust*PowerperCustomer.to('J'))
```

```python
print("Fraction %5.2e"%(num_cust*PowerperCustomer/Esol))
```
