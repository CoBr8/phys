```python
import numpy as np
from numpy import sqrt
import matplotlib.pyplot as plt
from datlib import clean_dat,ctree
import operator as op
from matplotlib import axes as ax
Re=6378000

def PC(lat, long):
    return np.array([np.cos(long)*np.cos(lat),np.sin(long)*np.cos(lat),np.sin(lat)])

def dA(lat1,long1,lat2,long2):
    lat1*=np.pi/180
    lat2*=np.pi/180
    long1*=np.pi/180
    long2*=np.pi/180
    
    p1=PC(lat1,long1)
    p2=PC(lat2,long2)
    cosang=sum(p1*p2)
    return(Re*np.arccos(cosang))

BClat=48.42505
BClong=-123.36647
avg_dict={}
new_dat=[]
master_dat=dict()
for crime in ctree:
    distance={}
    for x in clean_dat:
        if x["parent_incident_type"] == crime:
            new_dat.append(x) 
            if crime in distance.keys():       
                lat2=x['latitude']*pi/180;long2=x['longitude']*pi/180
                lat1=BClat*pi/180;long1=BClong*pi/180
                d=2*Re*arcsin(sqrt(sin((lat2-lat1)/2)**2+cos(lat1)*cos(lat2)*sin((long2-long1)/2)**2))
                distance[crime].append(d)
            else:
                distance[crime]=[]
    X=[dA(BClat,BClong,x['latitude'],x['longitude']) for x in new_dat]
    d_avg=sum(distance[crime])/len(distance[crime])
    avg_dict[crime]=d_avg
    f=plt.figure(figsize=(8,8))
    plt.hist(X,bins=[500*k for k in range(15)],alpha=0.5)
    plt.title(crime)
    plt.ylabel('Occourances')
    plt.xlabel('Distance from Bay Center')
    ax=f.add_subplot(111)
    textstr="""Average distance of \n"%s" \nfrom the Bay Centre: \n%d meters"""%(crime,int(d_avg))
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.3)
    plt.text(.5, 0.9, textstr,fontsize=12,verticalalignment='top', bbox=props,transform=ax.transAxes)
    plt.show()
avg_dict = sorted(avg_dict.items(), key=op.itemgetter(1))
pp.pprint(avg_dict)
```

```python
def timet(y0,y,v):
    t=(y-y0)/v
    return t

def position(y0,v,t):
    y=y0+v*t
    return y

def coltime(y0a,y0b,va,vb):
    t=(y0b-y0a)/(va-vb)
    return t

def vel(va_i,vb_i,ma,mb):
    from numpy import matrix
    '''
    Input:
    
        va_i := ball a initial velocity
        vb_i := ball b initial velocity
        ma   := mass of ball a
        mb   := mass of ball b
    
    Returns:
    
        va_f := ball a final velocity
        vb_f := ball b final velocity
        
    '''
    scalar_mult=1/(ma+mb)
    V=scalar_mult*matrix([[ma-mb,2*mb],[2*ma,mb-ma]])*matrix([va_i,vb_i]).T
    va_f=V[0];vb_f=V[1]
    return float(va_f),float(vb_f)
for mb in [10**i for i in range(1,3)]:
    va=0;vb=-1
    ma=1;
    ball_a=[1,1]
    ball_b=[2,1]
    time=[0,1]
    van,vbn=vel(va,vb,ma,mb)
    while vbn<0:
        dt=timet(ball_a[-1],0,van)
        time.append(time[-1]+timet(ball_a[-1],0,van))
        ball_a.append(position(ball_a[-1],van,dt))
        ball_b.append(position(ball_b[-1],vbn,dt))
        van=-van
        dt=coltime(ball_a[-1],ball_b[-1],van,vbn)
        time.append(time[-1]+timet(ball_a[-1],position(ball_b[-1],vbn,dt),van))
        ball_a.append(position(ball_a[-1],van,dt))
        ball_b.append(position(ball_b[-1],vbn,dt))
        van,vbn=vel(van,vbn,ma,mb)
    while abs(van)>abs(vbn):
        dt=timet(ball_a[-1],0,van)
        time.append(time[-1]+timet(ball_a[-1],0,van))
        ball_a.append(position(ball_a[-1],van,dt))
        ball_b.append(position(ball_b[-1],vbn,dt))
        van=-van
        if abs(van)<abs(vbn):
            break
        dt=coltime(ball_a[-1],ball_b[-1],van,vbn)
        time.append(time[-1]+timet(ball_a[-1],position(ball_b[-1],vbn,dt),van))
        ball_a.append(position(ball_a[-1],van,dt))
        ball_b.append(position(ball_b[-1],vbn,dt))
        van,vbn=vel(van,vbn,ma,mb)
    dt=1
    time.append(time[-1]+1)
    ball_a.append(position(ball_a[-1],van,dt))
    ball_b.append(position(ball_b[-1],vbn,dt))
    wall=np.zeros(len(ball_a))
    plt.figure(figsize=(8,8))
    plt.plot(ball_a,time,ball_b,time,wall,time,'k--')
    plt.show()
```
