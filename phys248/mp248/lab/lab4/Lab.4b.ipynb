```python
import numpy as np
import matplotlib.pyplot as plt
def fun(x,r=1):
    return np.sqrt(r**2-x**2)
n=30
x=np.linspace(0,1,n)
w=x[1]-x[0]
x_bar=np.linspace(0+w/2,1-w/2,n)
y=fun(x)
y_bar=fun(x_bar)
plt.bar(x_bar,y_bar,width=w,align='center')
plt.plot(x,y,'k--')
plt.show()
area=w*y_bar
area=area.sum()
print(area*4)
```

### Exercise 2: 
Now, use the Monte Carlo method to caculate $\pi$. How can that be done? 

#### Sketch
* Generate random points in a square.
* Decide if point is in circle with the same diameter as the side of the square.
* The ratio of the area of the circle to the area of the square is proportional to the ratio of the points in the circle to the points in the square.

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import random as r
from numpy import pi
import math as mp

n = 10000
x = r.uniform(-1,1,n)
y = r.uniform(-1,1,n)
a = 1
tot1=0
for i in x**2+y**2:
    if i<=1:
        tot1+=1  
tot2=len(x)

circ_x = np.linspace(-1,1,n)
y1  = np.sqrt( a**2 - circ_x**2 ) *  1
y2  = np.sqrt( a**2 - circ_x**2 ) * -1

plt.plot( x, y,'1',circ_x, y1,'r', circ_x, y2,'r')
plt.show()
print((tot1/tot2)*4)
```
